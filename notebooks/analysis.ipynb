# ERP Integration Log Analysis

This notebook provides a lightweight exploratory analysis of simulated ERP integration logs to validate data patterns before applying automated monitoring.


```python
import pandas as pd


df = pd.read_csv("../data/integration_logs.csv")
df.head()
```

```python
df["status"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["processing_time_ms"].hist(bins=50)
plt.title("Processing Time Distribution")
plt.xlabel("Processing Time (ms)")
plt.ylabel("Transaction Count")
plt.show()
```
